````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('particle_records_tensor.csv', header=None, names=['t', 'i', 'j', 'k', 'ux', 'uy', 'uz', 'pk_xx', 'pk_xy', 'pk_xz', 'pk_yx', 'pk_yy', 'pk_yz', 'pk_zx', 'pk_zy', 'pk_zz', 'qk_x', 'qk_y', 'qk_z'])
mean = data.groupby('t').mean()
````

````python
fig, ax1 = plt.subplots()
ax1.plot(mean.index, mean['ux'], 'g', label='$u_x$')
ax1.plot(mean.index, mean['uy'], 'b', label='$u_y$')
ax2 = ax1.twinx()
ax2.plot(mean.index, mean['uz'], 'r', label='$u_z$')
ax1.legend(loc='upper left')
ax2.legend()
````

````python
fig, ax1 = plt.subplots()
ax1.plot(mean.index, mean['pk_xz'], 'g', label='$p^k_{xz}$')
ax1.plot(mean.index, mean['pk_yz'], 'b', label='$p^k_{yz}$')
# ax1.plot(mean.index, mean['pk_zx'], label='$p^k_{zx}$')
# ax1.plot(mean.index, mean['pk_zy'], label='$p^k_{zy}$')
ax1.legend(loc='upper left')
ax2 = ax1.twinx()
ax2.plot(mean.index, mean['pk_xy'], 'r', label='$p^k_{xy}$')
# ax2.plot(mean.index, mean['pk_yx'], label='$p^k_{yx}$')
ax2.legend()
````

````python
fig, ax1 = plt.subplots()
ax1.plot(mean.index, mean['pk_xx'], 'g', label='$p^k_{xx}$')
ax1.plot(mean.index, mean['pk_yy'], 'b', label='$p^k_{yy}$')
ax2 = ax1.twinx()
ax2.plot(mean.index, mean['pk_zz'], 'r', label='$p^k_{zz}$')
ax1.legend(loc='upper left', framealpha=0.5)
ax2.legend(loc='upper right', framealpha=0.8)
````

````python
fig, ax1 = plt.subplots()
ax1.plot(mean.index, mean['qk_x'], 'g', label='$q^k_x$')
ax1.plot(mean.index, mean['qk_y'], 'b', label='$q^k_y$')
ax2 = ax1.twinx()
ax2.plot(mean.index, mean['qk_z'], 'r', label='$q^k_z$')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')
````

````python
Ts = pd.read_csv('particle_records_Ts.csv', header=None)
fig, ax1 = plt.subplots()
ax1.plot(Ts[0], Ts[1], label='$T$')
ax2 = ax1.twinx()
plt.xlabel('time (s)')
plt.ylabel('temperature (K)')
# linear regression using numpy
A = np.vstack([Ts[0], np.ones(len(Ts[0]))]).T
m, c = np.linalg.lstsq(A, Ts[1], rcond=None)[0]
ax2.plot(Ts[0], Ts[1]-m*Ts[0]-c, 'r', label='$T - T_{lin fit}$')
# plt.ylim(Ts[1].min(), Ts[1].max())
ax1.legend()
ax2.legend()
````
